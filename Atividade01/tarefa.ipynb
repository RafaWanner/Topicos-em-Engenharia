`https://insideairbnb.com/explore/`

```python
# Primeiro bloco: baixar arquivos do InsideAirbnb e salvar em ./data
import os
import re
import 
import requests
from urllib.parse import urljoin

data_dir = 'data'
os.makedirs(data_dir, exist_ok=True)

page_url = 'https://data.insideairbnb.com/china/hk/hong-kong/2025-09-23/data/listings.csv.gz'

if 'COLE_AQUI' in page_url:
    raise ValueError('Edite a variável page_url com a URL da página de dados da cidade')

resp = requests.get(page_url, timeout=30)
resp.raise_for_status()
html = resp.text

# Procura links para arquivos CSV, CSV.GZ ou ZIP na página
links = re.findall(r'href=["]([^"]+(?:csv|csvgz|zip))["]', html)
links = [urljoin(page_url, l) for l in links]

if not links:
    print('Nenhum arquivo CSV/ZIP encontrado na página. Verifique a URL.')
else:
    for url in links:
        fname = os.path.basename(url.split('?')[0])
        out_path = os.path.join(data_dir, fname)
        if os.path.exists(out_path):
            print(f'Já existe: {fname}, pulando.')
            continue
        print(f'Baixando {fname} ...')
        with requests.get(url, stream=True, timeout=60) as r:
            r.raise_for_status()
            with open(out_path, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
        print(f'Salvo em {out_path}')
print('Concluído.')
```
